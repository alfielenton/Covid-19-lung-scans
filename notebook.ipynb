```python
import json
import random
from matplotlib import pyplot as plt
from data_handling import DataConverter
```

```python
with open('training_results.json','r') as f:
    data = json.load(f)

epoch_losses = data['epoch loss']
forward_pass_losses = data['forward pass loss']
```

```python
plt.plot(epoch_losses['training'], color='blue', label='Training loss')

plt.legend()
plt.ylabel('Total loss over epoch per sample')
plt.xlabel('# Epochs')
plt.title('Model performance over epochs')
```

```python
plt.plot(forward_pass_losses['training'], linewidth=0.1, color='blue', label='Training loss')

plt.legend()
plt.ylabel('Forward pass loss per sample')
plt.xlabel('# forward pass')
plt.title('Model performance over forward passes')
```
